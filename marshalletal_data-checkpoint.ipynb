## Plotting of data from Marshall et al. 2025

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import xarray as xr
```

```python
df = pd.read_csv('Marshall_et_al_2024_Data_SAOD550nm_GlobalNH.csv', sep=',')
```

```python
# Hent ut årstall som heltall fra 'Date' (f.eks. "1250-01" → 1250)
df['Year'] = df['Date'].str.slice(0, 4).astype(int)

# Bruk en nøytral stil for å unngå grå bakgrunn, eksempelvis 'default'
plt.style.use('default')

# Sett opp figuren
plt.figure(figsize=(14, 7))

# Plotter alle kolonner unntatt 'Date' og 'Year'
for col in df.columns:
    if col not in ['Date', 'Year']:
        plt.plot(df['Year'], df[col], label=col)

# Definer x-aksen med bare hvert 100. år, starter på 1250
start_year = 1300
end_year = df['Year'].max()
plt.xticks(range(start_year, end_year + 1, 100))


volcano = [1257, 1448, 1448, 1815]
volcano_yvalue = [-0.01] * len(volcano)
plt.scatter(volcano, volcano_yvalue, color='red', marker='v', s=100)

# Tilpass akse og layout
plt.xlabel('Year')
plt.ylabel('SAOD550nm')
plt.title('Time series for all models')
plt.grid(True)
plt.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```

```python
files = [
    'prescribed_1257.nc',
    'prescribed_1458.nc',
    'prescribed_1815.nc',
    'prescribed_cntl.nc',
    'emissions_1257.nc',
    'emissions_1458_jan.nc',
    'emissions_1458_jul.nc',
    'emissions_1815.nc',
    'emissions_cntl.nc'
]

# Les inn hver fil som en xarray-dataset og lagre i en ordbok
data = {}

for file in files:
    key = file.replace('.nc', '')  # f.eks. 'prescribed_1257'
    data[key] = xr.open_dataset(file)

print(data['prescribed_1257'])
```

```python
ds=data['prescribed_1815']
zonal = ds['air_temperature'].mean(dim='realization')
baseline = zonal.isel(time=slice(0, 36)).mean(dim='time')
zonal_anomaly = zonal - baseline 
```

```python
ds_emission=data['emissions_1815']
zonal_e = ds_emission['air_temperature'].mean(dim='realization')
baseline_e = zonal.isel(time=slice(0, 36)).mean(dim='time')
zonal_anomaly_e = zonal_e - baseline_e 
```

```python
weights = np.cos(np.deg2rad(ds['latitude']))

# Globalt gjennomsnitt
global_temp = (ds['air_temperature'] * weights).sum(dim='latitude') / weights.sum()

# NH-maskering
NH_mask = ds['latitude'] > 0
weights_NH = weights.where(NH_mask)
NH_temp = (ds['air_temperature'] * weights_NH).sum(dim='latitude') / weights_NH.sum()

global_anom = global_temp - global_temp.isel(time=slice(0, 36)).mean(dim='time')
NH_anom = NH_temp - NH_temp.isel(time=slice(0, 36)).mean(dim='time')
```

```python
weights_e = np.cos(np.deg2rad(ds_emission['latitude_0']))

# Globalt gjennomsnitt
global_temp_e = (ds_emission['air_temperature'] * weights_e).sum(dim='latitude_0') / weights_e.sum()

# NH-maskering
NH_mask_e = ds_emission['latitude_0'] > 0
weights_NH_e = weights_e.where(NH_mask_e)
NH_temp_e = (ds_emission['air_temperature'] * weights_NH_e).sum(dim='latitude_0') / weights_NH_e.sum()

global_anom_e = global_temp_e - global_temp_e.isel(time=slice(0, 36)).mean(dim='time')
NH_anom_e = NH_temp_e - NH_temp_e.isel(time=slice(0, 36)).mean(dim='time')
```

```python
years_since = np.arange(global_anom.sizes['time']) / 12  # 12 måneder per år
years_since_e = np.arange(global_anom_e.sizes['time']) / 12
# Plot hver realization
for r in global_anom['realization'].values:
    plt.plot(years_since, global_anom.sel(realization=r), color='gray', alpha=0.3)

for r in global_anom_e['realization'].values:
    plt.plot(years_since_e, global_anom_e.sel(realization=r), color='mediumpurple', alpha=0.3)

# Gjennomsnittslinje
plt.plot(years_since, global_anom.mean(dim='realization'), color='black', label='Mean prescribed')
plt.plot(years_since_e, global_anom_e.mean(dim='realization'), color='slateblue', label='Mean emissions')

# Pynt
plt.xlabel("Years since eruption")
plt.ylabel("Temperature anomaly (K)")
plt.title("Global temperature anomaly – ensemble spread after 1815")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
years_since = np.arange(NH_anom.sizes['time']) / 12  # 12 måneder per år
years_since_e = np.arange(NH_anom_e.sizes['time']) / 12
# Plot hver realization
for r in NH_anom['realization'].values:
    plt.plot(years_since, NH_anom.sel(realization=r), color='gray', alpha=0.3)

for r in NH_anom_e['realization'].values:
    plt.plot(years_since_e, NH_anom_e.sel(realization=r), color='mediumpurple', alpha=0.3)

# Gjennomsnittslinje
plt.plot(years_since, NH_anom.mean(dim='realization'), color='black', label='Mean prescribed')
plt.plot(years_since_e, NH_anom_e.mean(dim='realization'), color='slateblue', label='Mean emissions')

# Pynt
plt.xlabel("Years since eruption")
plt.ylabel("Temperature anomaly (K)")
plt.title("NH temperature anomaly – ensemble spread after 1815")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
